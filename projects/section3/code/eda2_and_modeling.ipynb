```python
#선수들의 기록 데이터와 병합하기 위해 포지션을 더 큰 범위의 포지션으로 변경
mv_data.loc[mv_data.position.isin(['Centre-Forward', 'Left Winger', 'Right Winger', 'Second Striker']), 'position2'] = 'FW'
mv_data.loc[mv_data.position.isin(['Central Midfield', 'Left Midfield', 'Right Midfield']), 'position2'] = 'M'
mv_data.loc[mv_data.position == 'Defensive Midfield', 'position2'] = 'DMC'
mv_data.loc[mv_data.position == 'Attacking Midfield', 'position2'] = 'AM'
mv_data.loc[mv_data.position.isin(['Centre-Back', 'Right-Back', 'Left-Back']), 'position2'] = 'D'
mv_data.loc[mv_data.position == 'Goalkeeper', 'position2'] = 'GK'

mv_data['position2'].value_counts() 
```

```python
df = pd.read_csv('mv_data.csv',index_col = 0)
df['current_year'] = df['current_date'].apply(lambda x: x.split(',')[1].strip())
df['high_year'] = df['high_date'].apply(lambda x: x.split(',')[1].strip())
df.head()
```

```python
df['current_year'] = df['current_date'].apply(lambda x: x.split(',')[1].strip()) # 가장 최근의 몸값이 매겨진 년도
df['high_year'] = df['high_date'].apply(lambda x: x.split(',')[1].strip()) # 몸값이 가장 높았을 때의 년도
df['current_year'] = df['current_year'].astype(int)
df['high_year'] = df['high_year'].astype(int)
```

```python
#월을 숫자로 바꾸기
df['high_month'] = df['high_date'].apply(lambda x: x.split(' ')[0].strip())
df.loc[df['high_month'] == 'Jan','high_month'] = 1
df.loc[df['high_month'] == 'Feb','high_month'] = 2
df.loc[df['high_month'] == 'Mar','high_month'] = 3
df.loc[df['high_month'] == 'Apr','high_month'] = 4
df.loc[df['high_month'] == 'May','high_month'] = 5
df.loc[df['high_month'] == 'Jun','high_month'] = 6
df.loc[df['high_month'] == 'Jul','high_month'] = 7
df.loc[df['high_month'] == 'Aug','high_month'] = 8
df.loc[df['high_month'] == 'Sep','high_month'] = 9
df.loc[df['high_month'] == 'Oct','high_month'] = 10
df.loc[df['high_month'] == 'Nov','high_month'] = 11
df.loc[df['high_month'] == 'Dec','high_month'] = 12
df['high_month'].value_counts()
```

```python
#무조건 전시즌의 기록을 써야 Dataleakge를 피할 듯
#위에서 만든 년도 변수를 활용하여 몸값이 가장 높았을 때의 시즌 도출
df['seasons'] = None
for i in range(len(df)):
    df['seasons'][i] = str(df['high_year'][i]-1) + '-' + str(df['high_year'][i])
 
df.head()
```

```python
df['seasons'].value_counts().plot(kind = 'bar')
```

```python
df['birth_year'] = (df['birth'].apply(lambda x: x.split(' ')[2].strip())).astype(int) # 태어난 년도
df['height'] = (df['height'].apply(lambda x: x.replace(',','.').split(' ')[0])).astype(float)
df['current_mv'] = df['current_mv'].apply(lambda x: x[1:-1]).astype(float) 
df['high_mv'] = df['high_mv'].apply(lambda x: x[1:-1]).astype(float) 
df['high_month'] = df['high_month'].astype(int) 
df.head()
```

```python
# 지금은 한 행에 현 시즌, 가장 몸값이 높은 시즌의 데이터가 같이 있는데
# 두 시즌의 기록은 다를 것이기 때문에 이것을 분류한다.
# 현재 시즌의 몸값과 가장 높은 몸값의 시즌의 데이터를 분리해서 만들어준다.

df_current = df[['name', 'birth', 'height', 'country', 'position', 'position2', 'img_source', 'current_year', 'birth_year', 'current_mv']]
df_high = df[['name', 'birth', 'height', 'country', 'position', 'position2', 'img_source', 'high_year', 'birth_year', 'high_mv', 'seasons']]
print(df_current.shape)
print(df_high.shape)
display(df_current.head())
display(df_high.head())
```

```python
#가장 최근 몸값에는 20-21시즌의 데이터 추가. 아직 현 시즌이 안 끝났기 때문에 현 시즌 데이터 적용 X
df_current['seasons'] = '2020-2021'

#나이 데이터 추가
df_current['age'] = df_current['current_year'] - df_current['birth_year'] 
df_high['age'] = df_high['high_year'] - df_high['birth_year']

# 다시 합쳐주기 위해서 몸값, 년도 변수명 수정
df_current.rename(columns = {'current_mv':'market_values', 'current_year':'year'}, inplace = True)
df_high.rename(columns = {'high_mv':'market_values', 'high_year':'year'}, inplace = True) 

print(df_current.shape)
print(df_high.shape)
display(df_current.head())
display(df_high.head())
```

```python
mv_df = pd.concat([df_current, df_high])
print(sum(mv_df.duplicated())) # 500명 중 131명은 현재가 가장 높은 몸값이다.
mv_df.drop_duplicates(inplace = True) # 중복제거
print(sum(mv_df.duplicated()))
mv_df.head()
```

### 1. Linear Regression

```python
# 포지션별로 지표가 다르기 때문에 포지션별로 모델은 따로 만든다.
# 공격수 데이터
y_data_fw = fw_df_final['market_values']
X_train_fw, X_test_fw, y_train_fw, y_test_fw = train_test_split(fw_df_final.iloc[:,:13], y_data_fw, test_size = 0.2, random_state= 42)
display(X_train_fw.head())
display(y_train_fw.head())
```

```python
te = TargetEncoder(smoothing=1000) #smoothing이 1이 default 값으로 되어있다. 학습 데이터에 치우쳐진 평균을 전체 평균에 가깝도록 규제를 가한다. 1에 가까울 수록 더 큰 규제를 가하는 것.
X_train_fw_te = te.fit_transform(X_train_fw, y_train_fw)
X_test_fw_te = te.transform(X_test_fw)

display(X_train_fw_te.head())
display(X_test_fw_te.head())
```

```python
#미드필더 데이터
y_data_mid = mid_df_final['market_values']
X_train_mid, X_test_mid, y_train_mid, y_test_mid = train_test_split(mid_df_final.iloc[:,:13], y_data_mid, test_size = 0.5, random_state= 42)
display(X_train_mid.head())
display(y_train_mid.head())
```

```python
# 수비수 데이터
y_data_def = def_df_final['market_values']
X_train_def, X_test_def, y_train_def, y_test_def = train_test_split(def_df_final.iloc[:,:14], y_data_def, test_size = 0.2, random_state= 42)
display(X_train_def.head())
display(y_train_def.head())
```

```python
def calculate_score(data, pred):
    mae = mean_absolute_error(data,pred)
    mse = mean_squared_error(data, pred)
    rmse = mse ** 0.5
    r2 = r2_score(data, pred)
    return mae,mse,rmse,r2    
```

```python
#공격수 선형회귀
#과적합 발생
lm_fw = LinearRegression()
lm_fw.fit(X_train_fw_te, y_train_fw)

y_pred = lm_fw.predict(X_train_fw_te)

mae,mse,rmse,r2_train = calculate_score(y_train_fw, y_pred)
print('train 스코어',mae,mse,rmse,r2_train)

y_pred = lm_fw.predict(X_test_fw_te)

mae,mse,rmse,r2_test = calculate_score(y_test_fw, y_pred)
print('test 스코어',mae,mse,rmse,r2_test)

print('train, test r2 차이', r2_train - r2_test)
```

```python
#미드필더 선형회귀
#과적합 발생
lm_mid = LinearRegression()
lm_mid.fit(X_train_mid_te, y_train_mid)

y_pred = lm_mid.predict(X_train_mid_te)

mae,mse,rmse,r2_train = calculate_score(y_train_mid, y_pred)
print('train 스코어',mae,mse,rmse,r2_train)

y_pred = lm_mid.predict(X_test_mid_te)

mae,mse,rmse,r2_test = calculate_score(y_test_mid, y_pred)
print('test 스코어',mae,mse,rmse,r2_test)

print('train, test r2 차이', r2_train - r2_test)
```
