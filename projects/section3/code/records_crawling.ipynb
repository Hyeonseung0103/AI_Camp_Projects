```python
#시즌이나, 리그 바뀌면 네비게이트바 한번씩 눌러서 로딩한다음에 돌리기
pages = 1
i = 0
j = 1

try: 
    while pages < 29:
        print(i)
        if i > 9:
            i = 0
            j = 1
            print(f'{pages+1}로 넘어갑니다.')
            
            #summary 페이지 넘기기
            driver.find_elements_by_css_selector('#statistics-paging-summary > div > dl.listbox.right > dd:nth-child(4)')[0].click()
            time.sleep(1)
            
            #offensive 페이지 넘기기
            driver.find_elements_by_css_selector('.in-squad-detailed-view')[1].click()        
            time.sleep(1)
            driver.find_elements_by_css_selector('#statistics-paging-offensive > div > dl.listbox.right > dd:nth-child(4)')[0].click()

            #passing 페이지 넘기기
            driver.find_elements_by_css_selector('.in-squad-detailed-view')[2].click()
            time.sleep(1)
            driver.find_elements_by_css_selector('#statistics-paging-passing > div > dl.listbox.right > dd:nth-child(4)')[0].click()
            
            #defensive 페이지 넘기기
            driver.find_elements_by_css_selector('.in-squad-detailed-view')[0].click()
            time.sleep(1)
            driver.find_elements_by_css_selector('#statistics-paging-defensive > div > dl.listbox.right > dd:nth-child(4)')[0].click()

            pages +=1
            driver.find_elements_by_css_selector('li > a')[31].click()
            if pages >= 29:
                break

        games = driver.find_elements_by_css_selector('#player-table-statistics-body > tr')[i].get_attribute('innerText').split('\t')[0].split('\n')[-1].split('(')[0]
        mins = driver.find_elements_by_css_selector(f'#player-table-statistics-body > tr')[i].get_attribute('innerText').split('\t')[1]
        position = driver.find_elements_by_css_selector('td.col12-lg-2.col12-m-3.col12-s-4.col12-xs-5.grid-abs.overflow-text')[i].get_attribute('innerText').split('\n')[-1].split(',')[-1].split('(')[0].strip()
        
        if position == 'FW':
            name = driver.find_elements_by_css_selector('td.col12-lg-2.col12-m-3.col12-s-4.col12-xs-5.grid-abs.overflow-text')[i].get_attribute('innerText').split('\n')[1]
            team = driver.find_elements_by_css_selector('td.col12-lg-2.col12-m-3.col12-s-4.col12-xs-5.grid-abs.overflow-text')[i].get_attribute('innerText').split('\n')[-1].split(',')[0]
            driver.find_elements_by_css_selector('.in-squad-detailed-view')[1].click()
            contents = driver.find_elements_by_css_selector(f'#player-table-statistics-body > tr:nth-child({j})')[2].get_attribute('innerText').split('\t')[2:]
            df_fw = pd.concat([df_fw,pd.DataFrame([[name,league, seasons, team, position,games, mins]+contents], columns = df_fw.columns)])
            df_fw.reset_index(drop = True, inplace = True)
            j += 1
        elif position in ['M' ,'DMC', 'AM', 'Midfielder']:
            name = driver.find_elements_by_css_selector('td.col12-lg-2.col12-m-3.col12-s-4.col12-xs-5.grid-abs.overflow-text')[i].get_attribute('innerText').split('\n')[1]
            team = driver.find_elements_by_css_selector('td.col12-lg-2.col12-m-3.col12-s-4.col12-xs-5.grid-abs.overflow-text')[i].get_attribute('innerText').split('\n')[-1].split(',')[0]
            driver.find_elements_by_css_selector('.in-squad-detailed-view')[2].click()
            contents = driver.find_elements_by_css_selector(f'#player-table-statistics-body > tr:nth-child({j})')[3].get_attribute('innerText').split('\t')[2:]
            df_mid = pd.concat([df_mid,pd.DataFrame([[name,league, seasons, team, position,games, mins]+contents], columns = df_mid.columns)])
            df_mid.reset_index(drop = True, inplace = True)
            j += 1
        
        
        elif position in ['D', 'GK']:
            name = driver.find_elements_by_css_selector('td.col12-lg-2.col12-m-3.col12-s-4.col12-xs-5.grid-abs.overflow-text')[i].get_attribute('innerText').split('\n')[1]
            team = driver.find_elements_by_css_selector('td.col12-lg-2.col12-m-3.col12-s-4.col12-xs-5.grid-abs.overflow-text')[i].get_attribute('innerText').split('\n')[-1].split(',')[0]
            driver.find_elements_by_css_selector('.in-squad-detailed-view')[0].click()
            contents = driver.find_elements_by_css_selector(f'#player-table-statistics-body > tr:nth-child({j})')[1].get_attribute('innerText').split('\t')[2:]
            df_def = pd.concat([df_def,pd.DataFrame([[name,league, seasons, team, position,games, mins]+contents], columns = df_def.columns)])
            df_def.reset_index(drop = True, inplace = True)
            j += 1

        time.sleep(1)
        i += 1
        driver.find_elements_by_css_selector('li > a')[31].click()   


except:
    pass
    # df_fw.to_csv(f'df_fw_{seasons2}.csv')
    # df_mid.to_csv(f'df_mid_{seasons2}.csv')
    # df_def.to_csv(f'df_def_{seasons2}.csv')
    # df_fw.to_csv(f'{league2}/df_fw_serie_{seasons2}.csv')
    # df_mid.to_csv(f'{league2}/df_mid_serie_{seasons2}.csv')
    # df_def.to_csv(f'{league2}/df_def_serie_{seasons2}.csv')

            
# df_fw.to_csv(f'{league2}/df_fw_serie_{seasons2}.csv')
# df_mid.to_csv(f'{league2}/df_mid_serie_{seasons2}.csv')
# df_def.to_csv(f'{league2}/df_def_serie_{seasons2}.csv')




```
